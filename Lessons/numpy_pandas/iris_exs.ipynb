# Add a rank column to the iris DataFrame, ranking the petal length in an ascending order

```python
new_rank_df = df.sort_values(by=['petal_length']).reset_index(drop=True)
new_rank_df
```

```python
df['rank'] = df['petal_length'].rank(ascending=True)
df.set_index('rank', inplace=True)
df.sort_index()
```
