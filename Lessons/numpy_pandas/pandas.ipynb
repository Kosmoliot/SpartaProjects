```python
import pandas as pd
east = pd.Series([1000,1200,3400],index=['Q1','Q2','Q3'])
west = pd.Series([1100,1300,2400,3500],index=['Q1','Q2','Q3','Q4'])
```

```python
east
```

```python
df_region = pd.DataFrame({'East': east, 'West': west})
df_region
```

```python
df_region['North'] = [2000,3000,2500,4000]
df_region['South'] = [1500,2000,1500,4000]
df_region
```

```python
years = ['2016', '2017', '2018', '2019']
df_region['years'] = years
df_region
```

```python
df_region = df_region.set_index('years')
df_region
```

```python
new_df = df_region.reindex(['2017', '2018', '2019', '2020'])
new_df
```

```python
new_df = df_region.reindex(columns=['North', 'East', 'South', 'New'])
new_df
```

# methods bfill and ffill

```python
new_df.fillna(method='bfill')
```

```python
new_df.interpolate()
```

# dropna() with a tresh parameter indicating amount of 'na' values tolerated

```python
new_df.dropna(thresh=2)
```

```python
new_df.duplicated()
```

```python
new_df.drop_duplicates()
```
